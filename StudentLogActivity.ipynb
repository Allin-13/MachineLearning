```python
# Create activity_log.txt file in Colab
log_data = """\
S101 | Alice | LOGIN | 2024-03-01 | 09:00
S101 | Alice | SUBMIT_ASSIGNMENT | 2024-03-01 | 10:30
S101 | Alice | LOGOUT | 2024-03-01 | 11:00
S102 | Bob | LOGIN | 2024-03-01 | 09:15
S102 | Bob | SUBMIT_ASSIGNMENT | 2024-03-01 | 10:45
S102 | Bob | LOGIN | 2024-03-02 | 09:10
"""

with open("activity_log.txt", "w") as f:
    f.write(log_data)

print("activity_log.txt created successfully")
```

```python
import re
from collections import defaultdict

class Student:
    def __init__(self, student_id, name):
        self.student_id = student_id
        self.name = name
        self.activities = []

    def add_activity(self, activity):
        self.activities.append(activity)

    def summary(self):
        return (
            self.activities.count("LOGIN"),
            self.activities.count("SUBMIT_ASSIGNMENT")
        )

def read_logs(filename):
    pattern = re.compile(
        r'^(S\d+)\s*\|\s*(\w+)\s*\|\s*(LOGIN|LOGOUT|SUBMIT_ASSIGNMENT)\s*\|\s*(\d{4}-\d{2}-\d{2})\s*\|\s*(\d{2}:\d{2})$'
    )

    with open(filename, 'r') as file:
        for line in file:
            try:
                match = pattern.match(line.strip())
                if match:
                    yield match.groups()
            except:
                continue

students = {}
daily_activity = defaultdict(lambda: defaultdict(int))
login_status = defaultdict(int)

for sid, name, activity, date, time in read_logs("activity_log.txt"):
    if sid not in students:
        students[sid] = Student(sid, name)

    students[sid].add_activity(activity)

    # Store activity type per date
    daily_activity[date][activity] += 1

    if activity == "LOGIN":
        login_status[sid] += 1
    elif activity == "LOGOUT":
        login_status[sid] -= 1

with open("activity_report.txt", "w") as file:
    print("\nSTUDENT ACTIVITY REPORT\n")
    file.write("STUDENT ACTIVITY REPORT\n\n")

    for s in students.values():
        logins, submissions = s.summary()
        abnormal = "YES" if login_status[s.student_id] > 0 else "NO"

        report = (
            f"Student ID      : {s.student_id}\n"
            f"Name            : {s.name}\n"
            f"Total Logins    : {logins}\n"
            f"Submissions     : {submissions}\n"
            f"Abnormal Access : {abnormal}\n"
            "-----------------------------\n"
        )

        print(report)
        file.write(report)

    print("DAILY ACTIVITY STATISTICS")
    file.write("DAILY ACTIVITY STATISTICS\n")

    for date, activities in daily_activity.items():
        print(f"\nDate: {date}")
        file.write(f"\nDate: {date}\n")

        for act_type, count in activities.items():
            line = f"  {act_type} : {count}\n"
            print(line.strip())
            file.write(line)
```
